Data was split into two dataframes because certain stats weren't recorded pre modern era. The year the 3pt line was added to the present day and years after is usually consider the modern era and the years before are consider the pre modern era.

1) Modern era Years >= 1980

2) Pre Modern era Years <=1979

```python
modern_df=df[df['Year'] >= 1980]
modern_df.head()
```

```python
drop_co = ['3P','3PA','GS','3P%','ORB','DRB','STL','BLK','TOV','eFG%']

pre_mod_df=df[df['Year'] <= 1979] 

pre_mod_df = pre_mod_df.drop(drop_co, axis=1)

pre_mod_df.head()
```

## Data Wrangling:

Grouping the data by players

```python
player_mdf= modern_df.groupby('Player',as_index=False).sum().round(3)

player_pmdf =pre_mod_df.groupby('Player',as_index=False).sum().round(3) 
```

```python
#Checking to make sure the groupby worked
player_mdf.loc[player_mdf['Player'] == 'Michael Jordan*',:]
```

Had to change the totals to per game stats

Created a list of the columns to divide by total number of games to get the per game stats


**1) Modern Era**

```python
per_game = ['MP', 'FG', 'FGA', '3P', '3PA', '2P', '2PA', 'FT', 'FTA',
       'ORB', 'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS']

#For loop thats divides each column by the Games column
for stat in per_game:
    player_mdf[stat] = player_mdf[stat].div(player_mdf['G'])
    
    
    

# Create three separate list and ran a for loop to calculate the percentages for certain stats    
percent = ['FG%','3P%', '2P%', 'FT%']

made= ['FG','3P','2P','FT' ]

attempt=['FGA','3PA', '2PA', 'FTA']

 
for p,m,a in zip (percent,made,attempt):
    player_mdf[p]= player_mdf[m]/player_mdf[a]
    
    
    
    
#Used this formula to calculate the eFG% "(FG + 0.5 * 3P) / FGA". Formula was obtained from
# Basketball Reference

player_mdf['eFG%'] = (player_mdf['FG'] + 0.5 * player_mdf['3P'])/player_mdf['FGA']  
```

```python
player_mdf
```

**2) Pre Modern Era**

```python
per_game_p = ['MP', 'FG', 'FGA', '2P', '2PA', 'FT', 'FTA', 'TRB', 'AST',  'PF', 'PTS']

#For loop thats divides each column by the Games column
for stat in per_game_p:
    
    player_pmdf[stat] = player_pmdf[stat].div(player_pmdf['G'])
    
    

# Create three separate list and ran a for loop to calculate the percentages for certain stats    
percent_p = ['FG%','2P%', 'FT%']

made_p= ['FG','2P','FT' ]

attempt_p=['FGA','2PA', 'FTA']

 
for p,m,a in zip (percent_p,made_p,attempt_p):
    
    player_pmdf[p]= player_pmdf[m]/player_pmdf[a]
```

Dropped the Year and Age columns because they aren't necessary and rounded the data to 2 decimal places

## Creating the Hall of Fame column

```python
#Creating the HoF column and applying values of Yes/No. Yes if they were in the hall of fame and no if they weren't
player_mdf['HoF'] = player_mdf['Player'].apply(lambda x: 'Yes' if '*' in x else 'No')


player_pmdf['HoF'] = player_pmdf['Player'].apply(lambda x: 'Yes' if '*' in x else 'No')
```

```python
# Getting rid of the '*' in the Player column
player_mdf['Player'] = player_mdf['Player'].apply(lambda x: x.split('*')[0])


player_pmdf['Player'] = player_pmdf['Player'].apply(lambda x: x.split('*')[0])
```
